```python
def level1_feat(team,date):
    i = 0
    matches_no = 0
    runs_scored = 0
    runs_conceded = 0
    wins_no = 0
    wickets_taken = 0
    wickets_lost = 0
    balls_bowled = 0
    balls_batted = 0
    while date >= data_final['date'][i]:
        if data_final['winner'][i]==team:
            wins_no = wins_no + 1
        if data_final['team1'][i] == team:
            runs_scored = runs_scored + data_final['team1_bat'][i]
            matches_no = matches_no + 1
            balls_bowled = balls_bowled + data_final['team1_bowl'][i]
            balls_batted = balls_batted + data_final['team2_bowl'][i]
            runs_conceded = runs_conceded + data_final['team2_bat'][i]
            wickets_taken = wickets_taken + data_final['team1_wic'][i]
            wickets_lost = wickets_lost + data_final['team2_wic'][i]
        elif data_final['team2'][i] == team:
            runs_scored = runs_scored + data_final['team2_bat'][i]
            matches_no = matches_no + 1
            balls_bowled = balls_bowled + data_final['team2_bowl'][i]
            balls_batted = balls_batted + data_final['team1_bowl'][i]
            runs_conceded = runs_conceded + data_final['team1_bat'][i]
            wickets_taken = wickets_taken + data_final['team2_wic'][i]
            wickets_lost = wickets_lost + data_final['team1_wic'][i]
    wins_percent = wins_no/matches_no
    batting_run_rate = (runs_scored*6)/balls_batted
    bowling_econ_rate = (runs_conceded*6)/balls_bowled
    batting_avg = runs_scored/wickets_lost
    bowling_avg = runs_conceded/wickets_taken
    batting_wic_rate = wickets_lost/balls_batted
    bowling_strike_rate = wickets_taken/balls_bowled
    batting_index = batting_run_rate * batting_avg
    bowling_index = bowling_econ_rate * bowling_avg
    
    features = [wins_percent,batting_run_rate,bowling_econ_rate,batting_avg,bowling_avg,batting_wic_rate,bowling_strike_rate,batting_index,bowling_index]
    
    return features
```
